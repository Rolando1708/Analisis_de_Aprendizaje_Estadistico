```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import linear_model
import numpy as np
from sklearn.metrics import r2_score
```

```python
df = pd.read_csv('housing.csv')
```

```python
df = df[df.median_house_value!= df.median_house_value.max()]
```

```python
plt.scatter(df.median_income, df.median_house_value)
```

```python
y = df['median_house_value']
```

```python
# Total Rooms
X = df[['total_rooms']]
model= linear_model.LinearRegression().fit(X, y)
print(r2_score(y_true=y, y_pred=model.predict(X)))
print(model.coef_)
```

```python
# Median Income
X = df[['median_income']]
model= linear_model.LinearRegression().fit(X, y)
print(r2_score(y_true=y, y_pred=model.predict(X)))
print(model.coef_)
```

```python
# Total Bedroooms
df['total_bedrooms'] = df['total_bedrooms'].fillna(0)
X = df[['total_bedrooms']]
model= linear_model.LinearRegression().fit(X, y)
print(r2_score(y_true=y, y_pred=model.predict(X)))
print(model.coef_)
```

```python
# Median Age
X = df[['housing_median_age']]
model= linear_model.LinearRegression().fit(X, y)
print(r2_score(y_true=y, y_pred=model.predict(X)))
print(model.coef_)
```

```python
# Households
X = df[['households']]
model= linear_model.LinearRegression().fit(X, y)
print(r2_score(y_true=y, y_pred=model.predict(X)))
print(model.coef_)
```

```python
# Population
X = df[['population']]
model= linear_model.LinearRegression().fit(X, y)
print(r2_score(y_true=y, y_pred=model.predict(X)))
print(model.coef_)
```

```python
X = df[['median_income']]
model = linear_model.LinearRegression().fit(X,y)

print(model.intercept_)
print(model.coef_)
```
